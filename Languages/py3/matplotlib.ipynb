##### 1.03 Knowing the Difference Between bytes, str & unicode
    * In Python3, sequences are either BYTES (raw 8-bit values) or STRING (Unicode characters). Unicode values need to be 'encoded', while raw values need to be 'decoded'.
    * Try to keep sequences consitent with the UTF-8 Unicode encoding (8-bit representations of Unicode characters).  Use a pair of helper functions to do so.
    * Explicitly use binary modes (rb,wb) when working with files in Python3, since the default encoding format used  by open() is 'utf-8'.

```python
def to_bytes(value):
    """ takes a STRING/BYTES, returns a BYTES """
    if isinstance(value, str):
        value = value.encode('utf-8')
    return value
def to_string(value):
    """ takes a STRING/BYTES, returns a STRING """
    if isinstance(value, bytes):
        value = value.decode('utf-8')
    return value
```

##### 1.04 Write Helper Functions Instead of Complex Expressions
    Use helper functions whenever a logic is needed repeatedly.

```python
from urllib.parse import parse_qs

def get_value(values,key,default='<missing>'):
    found = values.get(key,[''])
    if found[0]:
        found = str(found[0])
    else:
        found = default
    return found


in_url = 'utm_source=facebook&utm_medium=social&utm_campaign=Build+Your+Career+with+ISA+International'

values = parse_qs(in_url, keep_blank_values=True)

print('Base URL:',get_value(values,'http://'))
print('Source:',get_value(values,'utm_source'))
print('Medium:',get_value(values,'utm_medium'))
print('Campaign:',get_value(values,'utm_campaign'))
```

##### 1.05 Know How to Slice Sequences
    * Slicing works for list, set, and tuple. It can be extended to any Python class that implements the __getitem__ and __setitem__ methods.
    * Slicing creates an entirely new list.
    * If defining use for 'n' elements of a list 'foo' whose size in unknown, IndexError can be avoided by slicing.
    Example: first_n_elements, last_n_elements = foo[:n], foo[-n:]

```python
import string

alphabets = [x for x in string.ascii_lowercase]

print('* All:',alphabets[:])
print('* First four:',alphabets[:4])
print('* Last four:',alphabets[-4:])
print('* Middle three:',alphabets[12:-11])
print('* All but last:',alphabets[:-1])
print('* Fifth onwards:',alphabets[4:])
print('* None but last three:',alphabets[-3:])
print('* Second onwards, but not last four:',alphabets[2:-4])
```
