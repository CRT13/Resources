```python
import wikipedia as wiki
```

__Query Wikipedia for pages with matching titles__

```python
wiki.search('Python')
```

__Link to title page__

```python
wiki.page('Python (programming language)').url
```

__Summary of title page__

```python
wiki.summary('Python (programming language)')
```

__Retrieve fullpage data for a title__

```python
page = wiki.page('Python (programming language)')
```

- __Extracting full-text__

```python
page.content
```

- __Extract links on page__

```python
page.references
```

- __Extract titles for links in the page__

```python
page.links
```

__Search for location using co-ordinates__

```python
station = wiki.geosearch(28.6427,77.2199)
print(station)
```

__Obtain (links to) all images for a title page__

```python
wiki.page(station).images
```

__Entire page in HTML__

```python
wiki.page(station).html()
```
